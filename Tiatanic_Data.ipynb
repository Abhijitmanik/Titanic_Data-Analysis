```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df=pd.read_csv("/content/Titanic_Data.csv")
```

```python
# Drop row with missing values
df_cleaned_rows = df.dropna()
print(df_cleaned_rows)
```

```python
# Drop columns with missing values
df_cleaned_cols = df.dropna(axis=1)

# Display the cleaned DataFrame
print(df_cleaned_cols)
```

```python
# Fill missing values with mean for numerical columns
df_filled = df.fillna(df.mean())

print(df_filled)
```

```python
#survival rate
survival_rate = df['Survived'].mean()
print(f"Overall Survival Rate: {survival_rate * 100:.2f}%")
```

```python
# Calculate the mean and standard deviation of the age of passengers:


# Assuming your dataset is loaded into a DataFrame called 'df'
age_mean = np.mean(df['Age'])
age_std = np.std(df['Age'])

print(f"Mean Age: {age_mean:.2f}")
print(f"Standard Deviation of Age: {age_std:.2f}")
```

```python
# Create a correlation matrix for numerical columns:

# Assuming your dataset is loaded into a DataFrame called 'df'
correlation_matrix = np.corrcoef(df[['Age', 'Fare', 'Survived']].dropna().T)
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Identify and count the unique values in the 'Pclass' column:

unique_classes = np.unique(df['Pclass'])
class_counts = np.unique(df['Pclass'], return_counts=True)

print("Unique Passenger Classes:", unique_classes)
print("Class Counts:", dict(zip(class_counts[0], class_counts[1])))
```

```python
# Group by passenger class and calculate survival rate
class_survival = df.groupby('Pclass')['Survived'].mean()

# Plot the results
class_survival.plot(kind='bar', color='skyblue')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#gender survival
gender_survival = df.groupby('Sex')['Survived'].mean()

gender_survival.plot(kind='bar', color=['lightblue', 'lightcoral'])
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Age Distribution of Passengers
plt.hist(df['Age'].dropna(), bins=20, color='lightcoral', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['Age'], df['Pclass'], c=df['Survived'], cmap='viridis', alpha=0.8)
plt.title('Scatter Plot of Age, Class, and Survival')
plt.xlabel('Age')
plt.ylabel('Passenger Class')
plt.colorbar(label='Survived')
plt.show()
```

```python
class_gender_survival = df.pivot_table(values='Survived', index='Pclass', columns='Sex', aggfunc='mean')

sns.heatmap(class_gender_survival, annot=True, cmap='Blues', fmt='.2f')
plt.title('Survival Rate by Class and Gender')
plt.xlabel('Gender')
plt.ylabel('Passenger Class')
plt.show()
```

```python
#  The embarkation point relate to the survival rate


# Group by embarkation point and calculate survival rate
embarkation_survival = df.groupby('Embarked')['Survived'].mean()

# Plot the results
embarkation_survival.plot(kind='bar', color='darkgreen')
plt.title('Survival Rate by Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#average fare paid by passengers in each class

# Group by passenger class and calculate average fare
average_fare_by_class = df.groupby('Pclass')['Fare'].mean()

# Plot the results
average_fare_by_class.plot(kind='bar', color='orange')
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show(0)
```

```python
# distribution of ticket prices

# Plot a histogram for the distribution of ticket prices
plt.hist(df['Fare'].dropna(), bins=30, color='salmon', edgecolor='black')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show
```
